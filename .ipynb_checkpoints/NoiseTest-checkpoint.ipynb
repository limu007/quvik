```python
from scipy import stats as st
import numpy as np
size=100
lev=300
img=st.poisson(lev).rvs(size=size*size).reshape(size,size)
```

```python
img22=(img[1:,1:]+img[1:,:-1]+img[:-1,1:]+img[:-1,:-1])/4
```

```python
from matplotlib import pyplot as pl
fcnt=pl.hist(img.ravel(),40)
pee=st.norm.fit(img.ravel())
sx=fcnt[1]
dx=sx[1]-sx[0]
pl.plot(sx,st.norm(*pee).pdf(sx)*size*size*dx)
```

```python
sx=fcnt[1]
dx=sx[1]-sx[0]
pl.plot(sx,st.norm(*pee).pdf(sx)*size*size*dx)
```

```python
from scipy import ndimage as nd
wee=nd.convolve1d(nd.convolve1d(img,[1,1,1],axis=1),[1,1,1],axis=0)/9
img33=wee[1:-1,1:-1]
fcnt=pl.hist(img33.ravel(),40)
```

```python
import numpy as np
psf=1.5
pixs=1.75
mx,my=np.mgrid[-4:4:80j,-4:4:80j]
zpsf=np.exp(-(mx**2+my**2)/2/psf**2)/(2*np.pi)/psf**2*0.1**2
zpsf.sum()
```

```python
from matplotlib import pyplot as pl
pl.imshow(zpsf,extent=[-4,4,-4,4])
pl.grid()
```

```python
from numpy import random as rd 
stab=3.5
niter=1000
sx=rd.randn(niter)*stab
sy=rd.randn(niter)*stab
dx,dy=0.1,0.1
foread=3
dchipx,dchipy=int(foread*pixs/dx),int(foread*pixs/dy)
cx,cy=len(mx)//2,len(mx[0])//2

drang=lambda c,s,d:[max(0,c-s),min(d,c+s)]
sumsub=lambda im,px,py:im[px[0]:px[1],py[0]:py[1]].sum()

allsim=[sumsub(zpsf,drang(int(cx-sx[j]/dx),dchipx,2*cx),drang(int(cy-sy[j]/dy),dchipy,2*cy)) for j in range(niter)]
```

```python
pl.hist(allsim,30);
```

```python
#pl.hist(np.sqrt(sx**2+sy**2),30);
def collef(stab=2.5,niter=1000,ptiles=[10,32,50]):
    sx=rd.randn(niter)*stab
    sy=rd.randn(niter)*stab
    allsim2=[sumsub(zpsf,drang(int(cx-sx[j]/dx),dchipx,2*cx),drang(int(cy-sy[j]/dy),dchipy,2*cy)) for j in range(niter)]
    pl.hist(allsim2,30);
    return np.percentile(allsim2,ptiles)
collef(2)
```
