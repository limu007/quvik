```python
import numpy as np

def sel_plot(name,unc=False,spl=25,extra='',reduce=3,exsel=None,nfilter='U',ncolor='navy',mstyle=None):
    osel=zimg.data['OBJECT']==name #object selection
    if np.iterable(exsel): csel=osel*exsel
    else: csel=osel*fsel
    ids1=np.where(csel)[0]
    t0=zimg.data['TRIGTIME'][ids1[0]]
    tobs=zimg.data['TSTART'][csel]-t0
    mag,emag=zimg.data['MAG'][csel],zimg.data['MAG_ERR'][csel]
    qsel=mag<99
    marker='.'
    if mstyle!=None: marker=mstyle[:1]
    lab=nfilter+' filter'
    if mstyle!=None: lab=name+" "+nfilter
    if unc:
        pl.errorbar(tobs[qsel][:spl],mag[qsel][:spl],emag[qsel][:spl],marker=marker,lw=0,elinewidth=1,color=ncolor,label=lab)
        if reduce>0:pl.errorbar(tobs[qsel][spl:],mag[qsel][spl:],emag[qsel][spl:],errorevery=reduce,marker=marker,lw=0,elinewidth=1,color=ncolor)
    else:
        pl.plot(tobs[qsel],mag[qsel],ncolor,label=lab)#,emag)weakest 
    if extra!="none":
        pl.title(name+extra)
    pl.ylabel(nfilter+' filter mag')
    pl.xlabel('time from trig [s]')
    pl.xscale("log")
    return [tobs[qsel],mag[qsel],emag[qsel]]
```

```python
sq1=sel_obj2[:10]
def make_cens(glist,fsel,first=5):
    # get some statistics on 
    cens=[]
    for s in glist:
        nsel=zimg.data['OBJECT'][fsel*mgood]==s
        val=zimg.data['MAG'][fsel*mgood][nsel][:first]
        if len(val)<3: val=np.array([0])
        cens.append([val.mean(),val.min(),val.max()])
    return np.array(cens).T
cens=make_cens(sel_obj2,fsel2)
```
