```python
fname='20deg50ms150nm_1_1.tif'
dimg1=np.array(PIL.Image.open(indir+fname))
dimg2=np.array(PIL.Image.open(indir+fname.replace('50ms','200ms')))
dimg1=dimg1[:2000,1000:2000].reshape(100,20,50,20)
pl.imshow(np.log(dimg1.mean(3).mean(1)),interpolation="nearest")
```

```python
dimg2=dimg2[:2000,1000:2000].reshape(100,20,50,20)
pl.plot(dimg1[dimg1>300],dimg2[dimg1>300],',')
```

```python
pl.hist(dimg1[dimg1>600],40);
```

```python
dint=(dimg1[dimg1>600]/60/4).astype(int)
from scipy import ndimage as nd
xval=np.r_[2:55]
yval=nd.median(dimg2[dimg1>600],dint,xval)
cnt=nd.sum(dimg2[dimg1>600],dint,xval)/nd.mean(dimg2[dimg1>600],dint,xval)
err=nd.standard_deviation(dimg2[dimg1>600],dint,xval)
```

```python
ind2,cov2=np.polyfit(xval*240,yval,1,w=np.sqrt(cnt)/err,cov=True)
ind2,np.sqrt(cov2.diagonal())
```

```python
pl.errorbar(xval*60*4,yval-np.polyval(ind2,xval*60*4),err/np.sqrt(cnt))
pl.title("resulduals of correl. fit @150nm")
```
