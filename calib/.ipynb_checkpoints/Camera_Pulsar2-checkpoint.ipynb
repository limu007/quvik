```python
def MAD_curve(fname,rat_range=[],mode="std",min_cnt=300,d=[20,20]):
    dimg1=np.array(PIL.Image.open(indir+fname))
    modimg1=dimg1[:2000,1000:2000].reshape(100,20,50,20).swapaxes(1,2)
    dimg1m=np.median(modimg1.reshape(100,50,400),2)
    difimg1=modimg1.reshape(100,50,400)-dimg1m[:,:,np.newaxis]
    smad=np.median(abs(difimg1),2)
    
    ssel=dimg1m>min_cnt
    ssel[-10:,:]=False
    ssel[:,:5]=False
    if len(rat_range)==2:
        rat1=smad/dimg1m
        ssel[rat1<=rat_range[0]]=False
        ssel[rat1>=rat_range[1]]=False

    from scipy import ndimage as nd
    sgend=dimg1m[ssel].max()
    if sgend>3e3: div=100
    else: div=int(sgend/20)
    labs=(dimg1m[ssel]/div).astype(int)
    imin=max(labs.min(),5)+1
    imax=min(labs.max(),60)-1
    if mode=='std':
        smad=np.sum(difimg1**2,2)/(d[0]*d[1]-1)
    rmean=nd.mean(smad[ssel],labs,np.r_[imin:imax])
    rcnt=nd.sum(smad[ssel],labs,np.r_[imin:imax])/rmean
    rstdv=nd.standard_deviation(smad[ssel],labs,np.r_[imin:imax])
    rsel=rcnt>3
    return np.r_[imin:imax][rsel]*div,rmean[rsel],rstdv[rsel],rcnt[rsel]
dd=MAD_curve(flst[6])
```

```python
for i in range(10,17):
    dd=MAD_curve(flst[i],rat_range=[0.03,0.06])
    pl.errorbar(*dd[:3])
    print(np.polyfit(dd[0][:10],dd[1][:10],1,w=1/dd[2][:10]))
```
