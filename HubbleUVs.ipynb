```python
uref={}
vref={}
i=0
for v in vis.MatchID:
    if v in uvids:
        uref[v]=uvids.index(v)
    vref[v]=i
    i+=1
```

```python
len(uref),len(uvids)
vis.W2_F606W[10]
```

```python
vcom=[vis.W2_F606W[vref[u]] for u in uref.keys()]
ucom=[fuv.W2_F300W[uref[u]] for u in uref.keys()]
ucom2=[vis.W2_F300W[vref[u]] for u in uref.keys()]
pl.plot(vcom,ucom,'+')
pl.plot(vcom,ucom2,'+')
pl.xlabel('606nm mag')
pl.ylabel('300nm mag')
```
